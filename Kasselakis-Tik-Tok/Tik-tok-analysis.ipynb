```python
# Function to clean and tokenize descriptions
def clean_and_tokenize(text):
    # Remove URLs, mentions, and special characters
    text = re.sub(r'http\S+|@\S+|#[^\s]+', '', text)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    # Convert to lowercase and tokenize
    tokens = text.lower().split()
    return tokens
```

```python
all_tokens = [token for desc in descriptions for token in clean_and_tokenize(desc)]
```

```python
# Remove common stop words
stop_words = set(['το', 'να', 'που', 'και', 'για', 'η', 'ο', 'τους', 'τον', 'την', 'δεν', 'των', 'μου', 'από', 'στο', 'του' ])

filtered_tokens = [token for token in all_tokens if token not in stop_words]
```

```python
word_freq = Counter(filtered_tokens)
```

```python
print(word_freq.most_common(20))
```

```python
# Function to extract hashtags
def extract_hashtags(text):
    return re.findall(r'#\w+', text)

# Extract hashtags from all descriptions
all_hashtags = [hashtag for desc in descriptions for hashtag in extract_hashtags(desc)]

# Count the frequency of each hashtag
hashtag_freq = Counter(all_hashtags)

# Display the most common hashtags
print("Most Common Hashtags:")
print(hashtag_freq.most_common(20))
```

```python
# Plot the most common hashtags
common_hashtags = dict(hashtag_freq.most_common(20))
plt.figure(figsize=(10, 5))
plt.bar(common_hashtags.keys(), common_hashtags.values())
plt.title('Most Common Hashtags')
plt.xticks(rotation=90)
plt.show()
```

```python
wordcloud_hashtag = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(hashtag_freq)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_hashtag, interpolation='bilinear')
plt.title('Word Cloud of Hashtags')
plt.axis('off')
plt.show()
```
