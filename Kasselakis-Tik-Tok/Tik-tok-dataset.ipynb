```python
from urllib.parse import parse_qs, urlparse
import hashlib
from pathlib import Path
import json

# If we've already added the listener, remove it
try:
    page.remove_listener('response', process_response)
except:
    pass

async def process_response(response):
    if response.ok and response.url.startswith("https://www.tiktok.com/api/"):
        try:
            parsed = urlparse(response.url)

            # Hash the query parameters and the response.text to make a unique filename
            m = hashlib.md5()
            m.update(parsed.query.encode('utf-8'))
            m.update((await response.text()).encode('utf-8'))
            filename = m.hexdigest()

            # the full path is based on the URL
            path = Path(parsed.path.lstrip("/")).joinpath(filename).with_suffix(".json")
            path.parent.mkdir(parents=True, exist_ok=True)

            print("Writing content to ", path)

            query = parse_qs(parsed.query)
            query = {k: v[0] for k, v in query.items()}

            content = {
                'url': response.url,
                'query': query,
                'headers': await response.all_headers(),
                'data': await response.json(),
            }

            path.write_text(json.dumps(content, indent=2))
        except Exception as e:
            print("Error processing response", e)

page.on("response", process_response)
```

```python
import glob
import json

# Get the list of JSON files in the specified directory
filenames = glob.glob("api/post/item_list/*.json")
print(filenames[:5])  # Print the first 5 filenames to verify

rows = []

# Extract video data from all files
for filename in filenames:
    try:
        with open(filename) as fp:
            content = json.load(fp)
            # Check if 'data' and 'itemList' keys exist
            if 'data' in content:
                if 'itemList' in content['data']:
                    rows.extend(content['data']['itemList'])
                else:
                    print(f"'itemList' not found in 'data' of file: {filename}")
                    print("Available keys in 'data':", content['data'].keys())
            else:
                print(f"'data' key not found in file: {filename}")
                print("Available keys:", content.keys())
    except Exception as e:
        print(f"Error processing file {filename}: {e}")

# Print the number of rows collected
print(f"Total rows collected: {len(rows)}")

# If rows are collected, normalize and print the first few entries
if rows:
    import pandas as pd
    pd.options.display.max_columns = None
    pd.options.display.max_colwidth = 75

    # Normalize JSON data into a DataFrame
    videos_df = pd.json_normalize(rows, sep='_')

    # Display the first 2 rows of the DataFrame
    print(videos_df.head(2))

    # Create URLs and insert them into the DataFrame
    urls = 'https://tiktok.com/@' + videos_df['author_uniqueId'] + '/video/' + videos_df['video_id']
    videos_df.insert(0, 'tiktok_url', urls)

    # Remove duplicates based on 'tiktok_url'
    videos_df = videos_df.drop_duplicates(subset=['tiktok_url'])

    # Print the total number of unique videos
    print(videos_df.shape[0], "total videos")

    # Save the DataFrame to a CSV file
    videos_df.to_csv("videos.csv", index=False)
else:
    print("No video data found.")
```

```python
import os
import yt_dlp
from yt_dlp.utils import ExtractorError, DownloadError
from urllib.error import HTTPError

df = pd.read_csv("videos.csv")

df.head()
```

```python
df.shape
```

```python
# In this case we're only downloading the first 40 using [:40]

for i, row in list(df.iterrows()):    
    url = row['tiktok_url']

    ydl_opts = {
        "outtmpl": os.path.join('videos', row['author_uniqueId'], "%(id)s.%(ext)s"),
        "ignore_errors": True,
        "quiet": True,
    }

    print("Downloading", url)

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        try:
            ydl.download([url])
        except (HTTPError, TypeError, ExtractorError, DownloadError) as e:
            # Catch urllib and yt-dlp errors when video not found
            print(
                f"Encountered error {e} when attempting to download url: {url}"
            )
```

```python
# Step 2: Get list of JSON files in the specified directory
filenames = glob.glob("api/post/item_list/*.json")
print(filenames[:10])  # Print the first 10 filenames to verify

rows = []

# Step 3: Extract comment data from all files
for filename in filenames:
    try:
        with open(filename) as fp:
            content = json.load(fp)
            # Ensure 'comments' key exists within 'data'
            if 'data' in content and 'comments' in content['data']:
                rows.extend(content['data']['comments'])
            else:
                print(f"'comments' key not found in 'data' of file: {filename}")
                print("Available keys in 'data':", content['data'].keys())
    except Exception as e:
        print(f"Error processing file {filename}: {e}")

# Step 4: Print the number of rows collected
print(f"Total rows collected: {len(rows)}")

# Step 5: Normalize JSON data into a DataFrame if rows are collected
if rows:
    comments_df = pd.json_normalize(rows, sep='_')

    # Display the first 2 rows of the DataFrame
    print(comments_df.head(2))

    # Step 6: Remove duplicates based on 'aweme_id' and 'cid'
    comments_df = comments_df.drop_duplicates(subset=['aweme_id', 'cid'])

    # Print the total number of unique comments
    print(comments_df.shape[0], "total comments")

    # Step 7: Save the DataFrame to a CSV file
    comments_df.to_csv("comments.csv", index=False)
else:
    print("No comment data found.")
```
